```python
import nltk
from nltk.chat.util import Chat, reflections
import requests
from bs4 import BeautifulSoup

# Fonction de recherche avec Mojeek
def search(query):
    url = f"https://www.mojeek.com/search?q={query}"
    headers = {'User-Agent': 'Mozilla/5.0'}

    try:
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            results = []
            for result in soup.find_all('a', class_='ob', limit=3):
                title = result.get_text()
                href = result['href']
                results.append(f"{title}: {href}")
            return "Here are the results found for your query:\n" + "\n".join(results) + "\nTake your time to visit these links and find what you need 😊" if results else "No results found."
        else:
            return "Failed to retrieve search results."
    except Exception as e:
        return f"Error: {str(e)}"

# Paires de questions et réponses
pairs = [
    [
        r"my name is (.*)",
        ["Hello %1, how can I assist you today?",]
    ],
    [
        r"hi|hello|hey",
        ["Hello!", "Hi there!", "Hey! How can I help you?"]
    ],
    [
        r"what is your name?",
        ["I am a simple chatbot created to assist you!",]
    ],
    [
        r"how are you?",
        ["I'm just a bunch of code, but thanks for asking!",]
    ],
    [
        r"quit",
        ["Bye! Take care!"]
    ],
    [
        r"(.*)distribution(.*)oil(.*)",
        ["Let me find some information about oil distribution for you.",]
    ],
    [
        r"(.*)pricing(.*)oil(.*)",
        ["Let me check the current pricing for oil.",]
    ],
    [
        r"(.*)delivery(.*)times?(.*)",
        ["I will look up the delivery times for you.",]
    ],
    [
        r"(.*)order(.*)status(.*)",
        ["I'll find out the status of your order.",]
    ],
    [
        r"(.*)fuel(.*)types(.*)",
        ["I will search for the types of fuel we offer.",]
    ],
    [
        r"(.*)invoices?(.*)payments?(.*)",
        ["Let me look up the information on invoices and payments.",]
    ],
    [
        r"(.*)customer(.*)support(.*)",
        ["I'll find the customer support information for you.",]
    ],
    [
        r"(.*)search for (.*)",
        ["Let me find some information for you on: %1"]
    ],
    [
        r"(.*)",
        ["I'm sorry, I don't understand that. Could you please rephrase your question?"]
    ]
]

def chatbot():
    print("Hi! I'm a simple chatbot. Type 'quit' to exit.")
    chat = Chat(pairs, reflections)
    while True:
        user_input = input("You: ")
        if user_input.lower() == 'quit':
            print("Bye! Take care!")
            break

        # Vérifier si l'entrée de l'utilisateur contient une requête de recherche
        if "search for" in user_input.lower():
            query = user_input.lower().replace("search for", "").strip()
            search_result = search(query)
            print("Chatbot:", search_result)
        else:
            response = chat.respond(user_input)
            if response is None:
                print("Chatbot: Sorry, I did not understand. Can you explain more? 🤔")
            else:
                print("Chatbot:", response)

if __name__ == "__main__":
    chatbot()
```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

model_name = "gpt2"  # You can change this to another model
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

def tokenize_input(text):
    input_ids = tokenizer.encode(text, return_tensors="pt")
    return input_ids

max_response_length = 50  # Adjust as needed

def generate_response(prompt, max_length=max_response_length, num_responses=2, temperature=0.7): # Changed num_responses to 2
    input_ids = tokenize_input(prompt)
    response_ids = model.generate(input_ids, max_length=max_length, num_return_sequences=num_responses, num_beams=2, no_repeat_ngram_size=2, temperature=temperature)

    responses = [tokenizer.decode(response, skip_special_tokens=True) for response in response_ids]
    return responses

    responses = [tokenizer.decode(response, skip_special_tokens=True) for response in response_ids]
    return responses

def generate_deterministic_response(prompt):
    return generate_response(prompt, temperature=0.2) # Pass temperature argument

def generate_random_response(prompt):
    return generate_response(prompt, temperature=0.8) # Pass temperature argument
```

```python
prompt = "Tell me a joke."
deterministic_responses = generate_deterministic_response(prompt)
random_responses = generate_random_response(prompt)

print("Deterministic Responses:")
for response in deterministic_responses:
    print(response)

print("\nRandom Responses:")
for response in random_responses:
    print(response)
```
